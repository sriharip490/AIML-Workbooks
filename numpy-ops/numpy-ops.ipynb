```python
# Import needed modules
import numpy as np
```

```python
# define a 2D array (3 rows and 2 columns)
data = np.array([[1, 2], [-3, 4], [5, 6]])
```

```python
print(data)
print("Shape of the array:", data.shape)
print("Data type of the array:", data.dtype)
```

```python
# Multiplication of the array with a scalar
A = data
B = 2 * A
print("Array A:\n", A)
print("Array B (A multiplied by 2):\n", B)
```

```python
# Multiplication of two arrays (element-wise)
# Note: Both A and B have the same shape, so we can perform element-wise multiplication
C = A * B
print("Element-wise multiplication of A and B (Array C):\n", C)
```

```python
# Multiplication of two matrices (dot product)
# For matrix multiplication, the number of columns in the first matrix must equal the number of rows in the second matrix. 
# Here, A is 3x2 and D is 2x3
D = np.array([[1, 0, 2], [0, 1, 3]])
E = np.dot(A, D)
print("Matrix D:\n", D)
print("Dot product of A and D (Array E):\n", E)
```

```python
# another form of matrix multiplication using the @ operator - same result as np.dot
F = A @ D
print("Matrix F (using @ operator for matrix multiplication):\n", F)
```

### Transpose and Inverse of Matrix

```python
# Transpose of a matrix
G = A.T
print("Transpose of Matrix A (Array G):\n", G)
```

```python
# Shape of Transpose of A
print("Shape of Transpose of A (Array G):", G.shape)
```

```python
# Inverse of a matrix A - Note: A must be a square matrix and non-singular to have an inverse.
# Since A is 3x2, it does not have an inverse. For demonstration, let's define a square matrix H and compute its inverse.
H = np.array([[1, -3, 5], [2, 4, 6], [0, 1, 2]])

HI = np.linalg.inv(H)
print("Inverse of Matrix H (Array HI):\n", HI)
```

```python
# Matrix multiplication of H and its inverse should yield the identity matrix
Identity = H @ HI
print("Product of H and its inverse (should be identity matrix):\n", Identity)
```

```python
# Determinant of a matrix
det_H = np.linalg.det(H)
print("Determinant of Matrix H:", det_H)
```

### Matrix Transformations

```python
# Matrix transformations - Example: Scaling a matrix by multiplying with a scalar. Each element 
# of the matrix is multiplied by the scalar value.
scalar = 8
scaled_H = scalar * H
print(f"Matrix H:\n", H)
print(f"Matrix H scaled by a factor of {scalar}:\n", scaled_H)
```

```python
# Matrix transformations - Example: Adding a scalar to each element of the matrix. 
# Each element of the matrix is increased by the scalar value.
scalar_add = 5
added_H = scalar_add + H
print(f"Matrix H:\n", H)
print(f"Matrix H with scalar {scalar_add} added to each element:\n", added_H)
```

```python
# Matrix transformations - Example: Scaling a matrix by multiplying with a vector
scale_vector = np.array([2, 3, -1])  # Scale x by 2, y by 3, and z by -1
scaled_H_vector = H @ scale_vector[:, np.newaxis]  # Scale each row of H by the corresponding element in the scale_vector
print(f"Matrix H:\n", H)
print(f"Matrix H scaled by vector {scale_vector}:\n", scaled_H_vector)
```

### Scaling - how it works with scale vector [2, 3, -1]
Columns:  X ------ Y ------ Z<br>
Row 0:  (1×2) + (-3×3) + (5×-1)  =  2 - 9 - 5  = -12<br>
Row 1:  (2×2) + ( 4×3) + (6×-1)  =  4 + 12 - 6 =  10<br>
Row 2:  (0×2) + ( 1×3) + (2×-1)  =  0 +  3 - 2 =   1<br>
- col x scaled by 2, 
- col y scaled by 3
- col z scaled by -1
